```javascript
const fs = require('fs');
const input = function (filename) {
  return fs.readFileSync(filename, 'utf8').split('\n').slice(0, -1);
}
```

```javascript
const ops = input("day8.txt");

// const ops = ["nop +0",
// "acc +1",
// "jmp +4",
// "acc +3",
// "jmp -3",
// "acc -99",
// "acc +1",
// "jmp -4",
// "acc +6"]
```

```javascript
const loop = function(ops) {
    let acc = 0;
    let i = 0;
    let visited = new Array();

    while (!visited.includes(i)) {
        if (i == ops.length) {
            console.log(i, acc);
            break;
        }
        visited.push(i);
        const [c, d] = ops[i].split(" ");
        if (c == "acc") {
            acc += Number(d);
            i += 1;
        }
        if (c == "jmp") {
            i += Number(d);
        }

        if (c == "nop") {
            i += 1;
        }
    }
    return acc;
}
```

```javascript
let mods = new Array();
for (let i in ops) {
    const c = ops[i];
    if (c.startsWith("nop")) {
        var d = c.replace("nop", "jmp");
    }
    if (c.startsWith("jmp")) {
        var d = c.replace("jmp", "nop");
    }
    let n = ops.slice();
    n[i] = d;
    mods.push(n);
}
```

```javascript
for (p of mods) {
    console.log(loop(p));
}
```
