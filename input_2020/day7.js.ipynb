```javascript
const fs = require('fs');
const input = function (filename) {
  return fs.readFileSync(filename, 'utf8').split('\n').slice(0, -1);
}
```

```javascript
const rules = input("day7.txt");

// const rules = ["light red bags contain 1 bright white bag, 2 muted yellow bags.",
//         "dark orange bags contain 3 bright white bags, 4 muted yellow bags.",
// "bright white bags contain 1 shiny gold bag.",
// "muted yellow bags contain 2 shiny gold bags, 9 faded blue bags.",
// "shiny gold bags contain 1 dark olive bag, 2 vibrant plum bags.",
// "dark olive bags contain 3 faded blue bags, 4 dotted black bags.",
// "vibrant plum bags contain 5 faded blue bags, 6 dotted black bags.",
// "faded blue bags contain no other bags.",
// "dotted black bags contain no other bags."];

// const rules = ["shiny gold bags contain 2 dark red bags.",
//  "dark red bags contain 2 dark orange bags.",
// "dark orange bags contain 2 dark yellow bags.",
// "dark yellow bags contain 2 dark green bags.",
// "dark green bags contain 2 dark blue bags.",
// "dark blue bags contain 2 dark violet bags.",
// "dark violet bags contain no other bags."]
```

```javascript
let bags = new Array("shiny gold");
```

```javascript
for (const b of bags) {
    for (const r of rules) {
        let [outer, inner] = r.split(" contain ");
        outer = outer.split(" ").slice(0, 2).join(" ");
        if (inner.includes(b) && !bags.includes(outer)) {
            bags.push(outer);
        }
    }
}
console.log(bags.length - 1);
```

```javascript
let parsed = new Map();
for (const rule of rules) {
    let [outer, inner] = rule.split(" contain ");
    const b = inner.split(", ");
    let items = new Map();
    for (const item of b) {
        const t = item.split(" ");
        if (!isNaN(Number(t[0]))) {
            items.set(t.slice(1,3).join(" "), Number(t[0]));
        }
    }
//     console.log(items);
    parsed.set(outer.split(" ").slice(0, 2).join(" "), items);
}
```

```javascript
// this is where everything goes haywire
const bag_count = function (bag) {
    let sum = 1;
    let inner = parsed.get(bag);
    for (const e of inner.entries()) {
        console.log(e);
        sum += e[1] * bag_count(e[0]);
    }
    return sum
}
```
