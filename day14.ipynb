```python
for x in range(25000):
    m = hashlib.md5()
    m.update(salt + str(x))
    hashes[x] = m.hexdigest().lower()
```

```python
triples = {}
quints = {}
for i in hashes.keys():
    for x in h:
        if x*3 in hashes[i]:
            triples[i] = x
        if x*5 in hashes[i]:
            quints[i] = x
```

```python
pprint.pprint(triples)
```

```python
pprint.pprint(quints)
```

```python
k = []
for n in quints.keys():
    for x in range(n, n-1000, -1):
        if x in triples.keys() and triples[x] == quints[n]:
            k.append(x)
```
