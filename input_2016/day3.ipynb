```python
def input(filename):
    with open(filename, 'r') as input:
        data = [x.strip() for x in input]

    return data
```

```python
def day3_check(triplet) -> bool:
    c = max(triplet)
    triplet.remove(c)
    a, b = triplet
    if a + b > c:
        return True
    else:
        return False
```

```python
triangles = input('day3.txt')
```

```python
htri = [list(map(int, x.split())) for x in triangles]
i = range(0, len(htri), 3)
vtri = [list(y) for x in i for y in zip(htri[x], htri[x + 1], htri[x + 2])]
```

```python
vtri
```

```python
p1 = len([x for x in htri[:] if day3_check(x)])
```

```python
p2 = len([x for x in vtri if day3_check(x)])
```

```python
print(p1, p2)
```
